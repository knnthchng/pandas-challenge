# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
schools_raw = "Resources/schools_complete.csv"
student_raw = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(schools_raw)
student_data = pd.read_csv(student_raw)

# Combine the data into a single dataset.  
all_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
all_data["grade"] = all_data["grade"].replace(['9th','10th','11th','12th'],[9,10,11,12])
all_data.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data["size"].sum()
student_count
```

```python
# Calculate the total budget
total_budget = sum(school_data["budget"])
total_budget
```

```python
# Calculate the average (mean) math score
dist_math_avg = all_data["math_score"].mean()
dist_math_avg
```

```python
# Calculate the average (mean) reading score
dist_read_avg = all_data["reading_score"].mean()
dist_read_avg
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
pass_math_count = all_data[(all_data["math_score"] >= 70)].count()["student_name"]
math_passrate = pass_math_count / float(student_count)
math_passrate
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
pass_read_count = len(all_data.loc[all_data["reading_score"] >= 70])
read_passrate = pass_read_count / float(student_count)
read_passrate
```

```python
# Use the following to calculate the percentage of students that passed math and reading
pass_all_count = all_data[
    (all_data["math_score"] >= 70) & (all_data["reading_score"] >= 70)
].count()["student_name"]
overall_passrate = pass_all_count /  float(student_count)
overall_passrate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
dist_summary = pd.DataFrame({"Total Schools": school_count,
                                "Student Population": student_count,
                                "Total Budget": total_budget,
                                "Average Math Score": dist_math_avg,
                                "Average Reading Score": dist_read_avg,
                                "Math Pass Rate": math_passrate,
                                "Reading Pass Rate": read_passrate,
                                "Overall Pass Rate": overall_passrate},
                               index = [0])

# Format and display dataframe
dist_summary.style.format({"Student Population" : "{:,}",
                           "Total Budget" : "${:,.2f}",
                           "Average Math Score" : "{:.1f}",
                           "Average Reading Score" : "{:.1f}",
                           "Math Pass Rate": "{:.1%}",
                           "Reading Pass Rate": "{:.1%}",
                           "Overall Pass Rate": "{:.1%}"})
```

## School Summary

```python
# Use the code provided to select the school type
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
# Calculate the total student count
school_pop = student_data.groupby("school_name")["student_name"].count()
```

```python
# Calculate the total school budget and per capita spending
school_budget = all_data.groupby(["school_name"]).mean()["budget"]
school_capita = school_budget / school_pop
# Create DataFrame for reference
school_percap = pd.Series(school_capita, name="per capita")
pd.merge(school_budget, school_percap, on="school_name")
```

```python
# Calculate the average test scores
per_school_math = all_data.groupby(["school_name"]).mean()["math_score"]
per_school_read = all_data.groupby(["school_name"]).mean()["reading_score"]
# Create DataFrame for reference
per_school_avg = pd.merge(per_school_math, per_school_read, on="school_name")
per_school_avg
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = len(per_school_math.loc[per_school_math >= 70])

# Calculate the number of schools with reading scores of 70 or higher
school_passing_read = len(per_school_read.loc[per_school_read >= 70])

# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_all = len(per_school_avg.loc[(per_school_avg["math_score"] >= 70) & (per_school_avg["reading_score"] >= 70)])

print(f"{school_passing_math} schools have an average math score of 70 or higher.")
print(f"{school_passing_read} schools have an average reading score of 70 or higher.")
print(f"{passing_all} schools have both average math and reading scores of 70 or higher.")
```

```python
# Use the provided code to calculate the passing rates
school_math_rate = all_data[all_data["math_score"] >= 70].groupby('school_name')['Student ID'].count() / school_pop
school_read_rate = all_data[all_data["reading_score"] >= 70].groupby('school_name')['Student ID'].count() / school_pop
school_all_pass = all_data[(all_data["math_score"] >= 70) & (all_data["reading_score"] >= 70)].groupby('school_name')['Student ID'].count() / school_pop
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
school_summary = pd.DataFrame({"School Type" : school_type,
                                   "Student Population" : school_pop,
                                   "Budget" : school_budget,
                                   "Budget Per Capita" : school_percap,
                                   "Average Math Score" : per_school_math,
                                   "Average Reading Score" : per_school_read,
                                   "Math Pass Rate": school_math_rate,
                                   "Reading Pass Rate": school_read_rate,
                                   "Overall Pass Rate": school_all_pass})

# Formatting
# https://towardsdatascience.com/style-pandas-dataframe-like-a-master-6b02bf6468b0
school_summary.style.format({"Student Population" : "{:,}",
                           "Budget" : "${:,.2f}",
                           "Budget Per Capita" : "${:,.2f}",
                           "Average Math Score" : "{:.1f}",
                           "Average Reading Score" : "{:.1f}",
                           "Math Pass Rate": "{:.1%}",
                           "Reading Pass Rate": "{:.1%}",
                           "Overall Pass Rate": "{:.1%}"})
                            
school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
spnd_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the Budget Per Capita ("Per Student Budget") 
schoolexpen_df = school_summary.copy()
schoolexpen_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
schoolexpen_df["Spending Ranges (Per Student)"] = pd.cut(schoolexpen_df["Budget Per Capita"],
                                                        spending_bins, labels=spnd_labels,
                                                        include_lowest=True)
schoolexpen_df = schoolexpen_df[["School Type","Student Population","Budget","Budget Per Capita",
                                "Spending Ranges (Per Student)", "Average Math Score",
                                "Average Reading Score", "Math Pass Rate",
                                "Reading Pass Rate","Overall Pass Rate"]]
schoolexpen_df
```

```python
#  Calculate averages for the desired columns. 
math_expen = schoolexpen_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
read_expen = schoolexpen_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
mathpass_expen = schoolexpen_df.groupby(["Spending Ranges (Per Student)"]).mean()["Math Pass Rate"]
readpass_expen = schoolexpen_df.groupby(["Spending Ranges (Per Student)"]).mean()["Reading Pass Rate"]
allpass_expen = schoolexpen_df.groupby(["Spending Ranges (Per Student)"]).mean()["Overall Pass Rate"]
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average Math Score":math_expen,
                                "Average Reading Score":read_expen,
                                "Math Pass Rate":mathpass_expen,
                                "Reading Pass Rate":readpass_expen,
                                "Overall Pass Rate":allpass_expen})
# Display results
spending_summary.style.format({"Average Math Score" : "{:.1f}",
                               "Average Reading Score" : "{:.1f}",
                               "Math Pass Rate": "{:.1%}",
                               "Reading Pass Rate": "{:.1%}",
                               "Overall Pass Rate": "{:.1%}"})
```

## Scores by School Size

```python
# Establish the bins.
pop_bins = [0, 1000, 2000, 5000]
pop_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

school_summary["School Size"] = pd.cut(schoolexpen_df["Student Population"],
                                                        pop_bins, labels=pop_labels,
                                                        include_lowest=True)
schoolsize = school_summary[["School Type","Student Population","School Size",
                                "Average Math Score","Average Reading Score", 
                                "Math Pass Rate","Reading Pass Rate","Overall Pass Rate"]]
schoolsize
```

```python
# Calculate averages for the desired columns. 
math_size = school_summary.groupby(["School Size"]).mean()["Average Math Score"]
read_size = school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
mathpass_size = school_summary.groupby(["School Size"]).mean()["Math Pass Rate"]
readpass_size = school_summary.groupby(["School Size"]).mean()["Reading Pass Rate"]
allpass_size = school_summary.groupby(["School Size"]).mean()["Overall Pass Rate"]
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({"Average Math Score":math_size,
                                "Average Reading Score":read_size,
                                "Math Pass Rate":mathpass_size,
                                "Reading Pass Rate":readpass_size,
                                "Overall Pass Rate":allpass_size})

# Display results
size_summary.style.format({"Average Math Score" : "{:.1f}",
                               "Average Reading Score" : "{:.1f}",
                               "Math Pass Rate": "{:.1%}",
                               "Reading Pass Rate": "{:.1%}",
                               "Overall Pass Rate": "{:.1%}"})
```
